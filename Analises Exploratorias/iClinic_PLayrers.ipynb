```
def RetornaGolsPlayerCasual(id_jogador):
  sql = 'select count(g.goal_id ) from goals g, matches m2 where g.match_id = m2.match_id'
  sql = sql + ' and  g.player_id = upper(\'' + id_jogador + '\')'
  sql = sql + ' and  m2.match_type = \'Casual\''
  cur_players.execute(sql)
  reg_goals = cur_players.fetchone()

  for reg in reg_goals:
    gols = reg
  
  return int(gols)
```

```
def RetornaGolsPlayerRanked(id_jogador):
  sql = 'select count(g.goal_id ) from goals g, matches m2 where g.match_id = m2.match_id'
  sql = sql + ' and  g.player_id = upper(\'' + id_jogador + '\')'
  sql = sql + ' and  m2.match_type = \'Ranked\''
  cur_players.execute(sql)
  reg_goals = cur_players.fetchone()

  for reg in reg_goals:
    gols = reg
  
  return int(gols)
```

# Inicio da Análise de Dados

```
# Criando dataset com dados da API
ds_players = [{"id": players['body'][item]['player_id'], 
               "nome": players['body'][item]["player_name"],
               "pais": players['body'][item]["country"],
               "gols_casual": 0,
               "gols_ranked": 0} for item in range(0,200)]
```

```
ds_players
```

```
# Criando um dataframe a partir do dataset 
df = pd.DataFrame(ds_players)
```

```
# Imprimindo o dataframe
df.info()
```

```
for index in range(0,199):
    df.loc[index,'gols_casual'] = RetornaGolsPlayerCasual(df.loc[index,'id'])
    df.loc[index,'gols_ranked'] = RetornaGolsPlayerRanked(df.loc[index,'id'])
```

```
df
```

```
# Fecha a conexão com o Banco de Dados
conexao.close()
```

# Melhores Player em partidas "Casual"

```
df_casual_sort = df.sort_values(['gols_casual'], ascending=False)
df_casual_sort
```

```
print('Melhor Player : ' + df_casual_sort['nome'][127] + ' com ' + str(df_casual_sort['gols_casual'][127]) + ' gols' )
```

# Melhores Player em partidas "Ranked"

```
df_ranked_sort = df.sort_values(['gols_ranked'],axis=0, ascending=False)
df_ranked_sort
```

```
print('Melhor Player : ' + df_ranked_sort['nome'][80] + ' com ' + str(df_ranked_sort['gols_ranked'][80]) + ' gols' )
```

```
# Agrupando os dados de gols  para o tipo de partida "Casual"
# Esté sendo classificado 4 tipos de níveis, de acordo com a quantidade de gols
df_gols_casual_ranges = df.copy()
bins=[0, 40, 60, 80, 100]

df_gols_casual_ranges['CasualGrupos'] = pd.cut(df_gols_casual_ranges['gols_casual'], 
                                   bins, 
                                   labels=["Nivel I", "Nivel II", "Nivel III", "Master"]) 

df_gols_casual_ranges
```

```
# Agrupando os dados de gols  para o tipo de partida "Ranked"
# Foi utilizado o mesmo critéria do "Casual" para a classificação dos níveis dos jogadores, de acordo com a quantidade de gols "Ranked"
df_gols_ranked_ranges = df_gols_casual_ranges.copy()
bins=[0, 40, 60, 80, 100]

df_gols_ranked_ranges['RankedGrupos'] = pd.cut(df_gols_ranked_ranges['gols_ranked'], 
                                   bins, 
                                   labels=["Nivel I", "Nivel II", "Nivel III", "Master"])

df_gols_ranked_ranges.head(50)
```
