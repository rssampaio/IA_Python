### Mini Projeto - Classificação - Identificação de Diabetes

```python
# Importação das libraries 
import numpy
import time
from sklearn.model_selection import train_test_split
from keras.optimizers import SGD
from keras.layers import Dense
from keras.models import Sequential
```

```python
#Exemplo de verificaçao de versões: uma library em específico ou todas.
!pip show keras
#!pip list
#!pip freeze
```

```python
# Carga dos dados
diabetes = numpy.loadtxt("diabetes.csv", delimiter = ",")
print(diabetes.shape)
```

```python
#Visualização do Array
diabetes
```

```python
#Definição das variáveis
X = diabetes[:,0:8] #independentes
Y = diabetes[:,8] #dependente
```

```python
# Definicao de Treino e Teste
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3)
```

```python
model = Sequential()
#Primeira camada com 16 neurônios com 8 variáveis de entrada, com função de ativação ReLU
model.add(Dense(16, input_dim = 8, kernel_initializer = 'uniform', activation = 'relu')) 

#Segunda Camada com 8 neurônios e com função de ativação ReLU
model.add(Dense(8, kernel_initializer = 'uniform', activation = 'relu'))

#Saída com 1 neurônio retornando se há câncer(tendência) ou não. 
#Alunos: Testem posteriormente com outra função de ativação.
model.add(Dense(1, activation = 'sigmoid', kernel_initializer = 'uniform'))
```

É uma estrutura totalmente conectada. Lembre-se que a função de ativação determina a saída de cada neurônio.

```python
# Compilando o modelo
model.compile(loss = 'binary_crossentropy', optimizer = SGD(lr = 0.01, momentum = 0.8, nesterov = True), metrics = ['accuracy'])

'''
Momentum incorpora as propriedades da atualização de pesos anterior forçando a atualização dos pesos no mesmo sentido quando ocorre a diminuição do erro.
'''
```

```python
# Execução do o modelo e validação nos mesmos dados(treino)
iniProc = time.time()
model.fit(X, Y, epochs = 150, batch_size = 10)
print("Execução durou %2.f segundos " % (time.time() - iniProc))
del iniProc
```

```python
# Compila o modelo com o Adam.
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
# Execução do o modelo e validação nos mesmos dados(treino)
iniProc = time.time()
model.fit(X, Y, epochs = 150, batch_size = 10)
print("Execução durou %2.f segundos " % (time.time() - iniProc))
del iniProc
#Dependendo das configurações do seu computador/notebook, pode demorar um pouco mais.
```

```python
#Identificar os melhores pesos
model.get_weights()
```

```python
#Resumo da Rede
model.summary()
```

```python
# Avaliação do Modelo
scores = model.evaluate(X, Y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
#Chamamos o modelo e solicitamos validação em um conjunto separado(no de Validação).
model.fit(X, Y, validation_split = 0.30, epochs = 150, batch_size = 10)
```

```python
# Chamamos o modelo e aplicamos validação em um conjunto de dados de teste
model.fit(X_train, y_train, validation_data=(X_test,y_test), epochs = 150, batch_size = 10)
```
